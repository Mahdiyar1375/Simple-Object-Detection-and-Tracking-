```python
from norfair import Detection, Tracker, Video, draw_tracked_objects
import cv2
from sahi.utils.cv import read_image
from sahi.predict import get_prediction, get_sliced_prediction, predict
from sahi.auto_model import AutoDetectionModel
from ultralytics import YOLO
import numpy as np
import norfair
```

```python
video = Video(input_path=r"C:\bus_top_view2.mp4")
tracker = Tracker(distance_function="euclidean", distance_threshold=20)
```

```python
fourcc = cv2.VideoWriter_fourcc('F','M','P','4')
out = cv2.VideoWriter(r'D:/output_tracked.mp4', fourcc, 60.0, (1920,1080))
```

```python
detection_model = AutoDetectionModel.from_pretrained(
    model_type='yolov8_obb',
    model_path=r"C:\Users\Mahdiyar\runs\obb\yolo_new_app9\weights\best.pt",
    confidence_threshold=0.75,
    device="cuda:0",  # or 'cuda:0'
)
```

```python
def yolo_detections_to_norfair_detections(yolo_detections):
    
    norfair_detections = []
    for res in yolo_detections:
        xmin, ymin, xmax, ymax = res.bbox.to_xyxy()
        center_x = (xmin + xmax) / 2
        center_y = (ymin + ymax) / 2
        centroid = np.array([center_x, center_y])
        scores = np.array([res.score.value])
        norfair_detections.append(Detection(points=centroid,scores=scores,label=int(res.category.id)))
        

    return norfair_detections
```

```python
i = 0
for frame in video:
    i = i + 1
    result = get_sliced_prediction(
    frame,
    detection_model,
    slice_height=640,
    slice_width=640,
    overlap_height_ratio=0.2,
    overlap_width_ratio=0.2
    )
    detections = yolo_detections_to_norfair_detections(result.object_prediction_list)
    tracked_objects = tracker.update(detections=detections)

    norfair.draw_points(frame, tracked_objects,draw_ids=False)
    #video.write(frame)
    out.write(frame)


#cap.release()
out.release()
cv2.destroyAllWindows()
```
